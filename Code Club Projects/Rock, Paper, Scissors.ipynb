```
from random import randint
```

Solution to the original problem

```
player = input('rock (r), paper(p) or scissors (s)?')

chosen = randint(1,3)
#print (chosen)

if chosen == 1:
  computer = 'r'
elif chosen == 2:
  computer = 'p'
else:
  computer = 's'
  
print (player, 'vs', computer)

if player==computer:
  print('DRAW!')
elif (player=='r' and computer=='s') or \
     (player=='p' and computer=='r') or \
     (player=='s' and computer=='p'):
  print('Player wins!')
else:
  print('Computer wins!')
 
```

Rock Paper Scissors Lizard Spock game

```
player = input('rock (r), paper(p), scissors (s), lizard (l) or spock (k)?')

chosen = randint(1,5)

choices = ['r','p','s','l','k']
computer = choices[chosen-1]
  
print (player, 'vs', computer)

if player==computer:
  print('DRAW!')
elif (player=='r' and (computer=='s' or computer=='l')) or \
       (player=='p' and (computer=='r' or computer=='z')) or \
       (player=='s' and (computer=='p' or computer=='l')) or \
       (player=='l' and (computer=='k' or computer=='p')) or \
       (player=='k' and (computer=='s' or computer=='r')):
  print('Player wins!')
else:
  print('Computer wins!')
```

Keep a continuous game and a scoore

```
keep_game = 'y'
player_score = 0
computer_score = 0

while (keep_game=='y' or keep_game =='yes'):
  player = input('rock (r), paper(p), scissors (s), lizard (l) or spock (k)?')

  chosen = randint(1,5)

  choices = ['r','p','s','l','k']
  computer = choices[chosen-1]

  print (player, 'vs', computer)

  if player==computer:
    print('DRAW!')
  elif (player=='r' and (computer=='s' or computer=='l')) or \
       (player=='p' and (computer=='r' or computer=='z')) or \
       (player=='s' and (computer=='p' or computer=='l')) or \
       (player=='l' and (computer=='k' or computer=='p')) or \
       (player=='k' and (computer=='s' or computer=='r')):
    print('Player wins!')
    player_score+=1
  else:
    print('Computer wins!')
    computer_score+=1
    
  print_string = 'Current score is: Player '+str(player_score)+', Computer '+ \
                 str(computer_score)+' do you want to play again? (y/n)'  
  keep_game = input(print_string)
```

Game where the first player to get to three wins the game

```
player_score = 0
computer_score = 0
choices = ['r','p','s','l','k']

while (player_score<3 and computer_score<3):
  player = input('rock (r), paper(p), scissors (s), lizard (l) or spock (k)?')

  if (player in choices):
    chosen = randint(1,5)

    computer = choices[chosen-1]

    print (player, 'vs', computer)

    if player==computer:
      print('DRAW!')
    elif (player=='r' and (computer=='s' or computer=='l')) or \
         (player=='p' and (computer=='r' or computer=='z')) or \
         (player=='s' and (computer=='p' or computer=='l')) or \
         (player=='l' and (computer=='k' or computer=='p')) or \
         (player=='k' and (computer=='s' or computer=='r')):
      print('Player wins!')
      player_score+=1
    else:
      print('Computer wins!')
      computer_score+=1
    
if player_score>computer_score:
  print ('Player is the winner!')
else:
  print ('Computer is the winner!')
```
