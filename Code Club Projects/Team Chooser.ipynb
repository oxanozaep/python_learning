```
players = ['Harry', 'Hermione', 'Neville', 'Ginny']
```

```
print ("Initial players:", players)

while len(players)>0:
  playerA = choice(players)
  print ("Chosen for team A:", playerA)
  teamA.append(playerA)
  print ("Team A:", teamA)
  players.remove(playerA)
  print ("Players left:", players)

  playerB = choice(players)
  print ("Chosen for team B:", playerB)
  teamB.append(playerB)
  print ("Team B:", teamB)
  players.remove(playerB)
  print ("Players left:", players)
```

### Dividing teams into a number specified by user

```
players = ['Harry', 'Hermione', 'Neville', 'Ginny', 'Ron', 'Draco', 'Luna']
team_names = ['Alligators', 'Gorillas', 'Eagles', 'Pythons', 'Wasps', \
              'Panthers']
```

```
keep_asking = True

while keep_asking:
  team_number = input('Insert the number of teams:')
  try: 
    isinstance(int(team_number), int)
    if int(team_number)>len(team_names):
      print ("The number of teams can't be higher than the number of names")
    else:
      keep_asking = False
  except:
    print ("You must enter an integer")
    
team_number = int(team_number)
print ("\nNumber of teams chosen:", team_number)
```

```
teams_created = []

for i in range(team_number):
  new_team = choice(team_names)
  team_names.remove(new_team)
  teams_created.append(new_team)
  
print ("Teams created:", teams_created)
```

```
player_lists = [[] for i in range(team_number)]

while len(players)>0:
  for i in range(team_number):
    player = choice(players)
    players.remove(player)
    player_lists[i].append(player)

    if players == []:
      break
      
for i in range(team_number):
  print ("Team", teams_created[i],player_lists[i])
```
