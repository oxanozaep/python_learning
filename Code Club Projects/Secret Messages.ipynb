### Base project

```
alphabet = 'abcdefghijklmnopqrstuvwxyz'
```

```
key = int(input('Please enter a key:'))
```

```
message = input('Please enter a message:')
newMessage = ""

for character in message:
  if character in alphabet:
    newPosition = (alphabet.find(character) + key) % len(alphabet)
    newMessage += alphabet[newPosition]
  else:
    newMessage += character
  
print ('The new message is',newMessage)
```

### Friendship calculator

```
names = []

names.append(input('Please enter the first name:'))
names.append(input('Please enter the second name name:'))
```

```
# Points for character similarity

similarityPoints = 0

for character in names[0]:
  if character in names[1]:
    similarityPoints += 1
    
for character in names[1]:
  if character in names[0]:
    similarityPoints += 1

similarityPoints = similarityPoints / (len(names[0]) + len(names[1]))

# Points for length similarity

if len(names[0]) == len(names[1]):
  lengthPoints = 1
else:
  lengthPoints = 1 - (abs(len(names[0]) - len(names[1]))/10)
  lengthPoints = max(0, lengthPoints)

  
finalPoints = (similarityPoints + lengthPoints) / 2
```

```
if finalPoints > 0.75:
  print ("Your names have a similarity index of an amazing {0:.2%}!! Congrats!!".format(finalPoints))
elif finalPoints > 0.5:
  print ("Your names have a very decent similarity index of {0:.2%}".format(finalPoints))
elif finalPoints > 0.25:
  print ("Your names have a rather poor similarity index of {0:.2%}".format(finalPoints))
else:
  print ("Your names are not similar at all based on a similarity index of {0:.2%}".format(finalPoints))
```

### File encryption

```
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```
startingString = str(uploaded['players.txt'])
startingString = startingString.replace('b\'', '')
startingString = startingString.replace('\'', '')
startingString = startingString.replace('\\r\\n', '')
startingString = startingString.replace('\\t', ',')
listToEncrypt = startingString.split(",")
print (listToEncrypt)
```

```
enrypted_list = []
for element in listToEncrypt:
  element = element.lower()
  newElement = ""
  for character in element:
    if character in alphabet:
      newPosition = (alphabet.find(character) + key) % len(alphabet)
      newElement += alphabet[newPosition]
    else:
      newElement += character
  
  enrypted_list.append(newElement)
```

```
with open('encrypted_file.txt', 'w') as f:
  for item in enrypted_list:
    f.write("%s\n" % item)

files.download('encrypted_file.txt')
```
